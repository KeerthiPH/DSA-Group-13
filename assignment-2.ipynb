```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('diabetes_model.csv')
```

# Count plot of Outcome

```python
ax = sns.countplot(data['Outcome'])
for p in ax.patches:
    ax.annotate(p.get_height() , (p.get_x()+0.1, p.get_height()-50))
```

# effect of age on Outcome using Boxplot

```python
sns.boxplot(x = 'Outcome' , y= 'Age' ,data = data)
```

Analysis: As age increses there is more chance to be diabetic

# chances of getting diabetes if it is common in family, using the Daiabetes Pedigree Function

```python
sns.boxplot(x = 'Outcome' , y= 'DiabetesPedigreeFunction' ,data = data)
plt.title(" DiabetesPedigree v/s Outcome")
```

# relationship with Insulin & Glucose and group the datapoints by Outcome.

```python
sns.scatterplot( x = data['Insulin'], y = data['Glucose'] , hue=data['Outcome'])
plt.show()
```

# heatmap on Correlation of features in the data.


```python
sns.set(rc={'figure.figsize':(12,8)})
sns.heatmap(data.corr(),annot=True)
```
